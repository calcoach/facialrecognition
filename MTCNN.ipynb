```
from mtcnn.mtcnn import MTCNN
from google.colab.patches import cv2_imshow
from datetime import datetime
import cv2
 
# extract a single face from a given photograph
def extract_face(image, required_size=(160, 160)):

	 detector = MTCNN()
	# detect faces in the image
	 results = detector.detect_faces(image)
	# extract the bounding box from the first face
	 return results
   

image = cv2.imread('/content/foto7.jpg')

now = datetime.now()
current_time = now.strftime("%H:%M:%S")
print("Current Time =", current_time)
faces = extract_face(image)

now = datetime.now()
current_time = now.strftime("%H:%M:%S")
print("Current Time =", current_time)

for i, face in enumerate (faces):
  x, y, w, h = faces[i]['box']
  cv2.rectangle(image, (x, y), (x + w, y + h), (255, 200, 0), 2)

cv2_imshow(image)



```

```
import cv2 
from faced import FaceDetector
from faced.utils import annotate_image
from datetime import datetime
from google.colab.patches import cv2_imshow


face_detector = FaceDetector()

now = datetime.now()
current_time = now.strftime("%H:%M:%S")
print("Current Time =", current_time)

img = cv2.imread('angie.jpg')

rgb_img = cv2.cvtColor(img.copy(),cv2.COLOR_BGR2RGB)

bboxes = face_detector.predict(rgb_img)

# Use this utils function to annotate the image.
ann_img = annotate_image(img, bboxes)

now = datetime.now()
current_time = now.strftime("%H:%M:%S")
print("Current Time =", current_time)

# Show the image
cv2_imshow(ann_img)

```

```
from mtcnn.mtcnn import MTCNN
import cv2


def extractFace(frame):

  detector = MTCNN()

  return detector.detect_faces(frame)


VIDEO = "/content/drive/My Drive/Videos/obama.mp4"

fourcc = cv2.VideoWriter_fourcc(*'XVID')
cap = cv2.VideoCapture(VIDEO)

out = cv2.VideoWriter('with10frames.avi', fourcc, 20.0, (int(cap.get(3)),int(cap.get(4))))
ret = True
ret , frame = cap.read()
faces  = extractFace(frame)
count_frame = 1
while ret:

  ret , frame = cap.read()

  count_frame = count_frame+1

  if(count_frame % 10 == 0 ):
     faces = extractFace(frame)

  if (count_frame % 20 == 0):
    print('{}'.format(count_frame/20) + ' segundos procesados')

  for i, feature in enumerate(faces):
     (x,y,w,h) = faces[i]['box']

     if(faces[i]['confidence']>0.95):
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 200, 0), 2)

     if count_frame % 10 == 0:
       print(faces[i]['confidence'])
     

  out.write(frame)
   





```
